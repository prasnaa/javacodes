```python
x_feature = df[['year']]
y_feature = df['per capita income (US$)']
```

```python
plt.xlabel('year')
plt.ylabel('per capita')
plt.scatter(x_feature,y_feature)
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(x_feature.shape,y_feature.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_feature,y_feature,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict([[1980]])
```

```python
plt.scatter(x_feature,y_feature)
plt.plot(X_train,model.predict(X_train),color='blue')
plt.xlabel('CGPA')
plt.ylabel('package')
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error
mse = root_mean_squared_error(y_pred=y_predict,y_true=y_test)
```

```python
mse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_predict)
print("R^2")
```
