```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Data = pd.read_csv(r"C:\Users\pdhau\OneDrive\Desktop\Third semester\AI AND ML\Student_Performance.csv")
```

```python
Data
```

```python
X = Data.drop(columns=['Performance Index'],axis=1)
y = Data['Performance Index'] 
```

```python
X.head()
```

```python
X = pd.get_dummies(X,drop_first=True,dtype=int)
X.head()
```

```python
from sklearn .model_selection import train_test_split;
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=20)
```

```python
#Standarizing the dataset with standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train= scaler.fit_transform(X_train)
x_test= scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
regression.fit(X_train,y_train)
```

```python
y_pred = regression.predict(X_test) 
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae = mean_absolute_error(y_test ,y_pred)
mse = mean_squared_error(y_test ,y_pred=y_pred)
score = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Mean square error is {mse}")
print(f"Mean absolute error is {mae}")
print(f"r2 score is {score}")
print(f"root mean square error is {rmse}")
```
